```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("202108_202208_주민등록인구및세대현황_월간.csv", encoding="cp949")
print(data)
```

```python
dataDict = {}
for i in range(1, len(data)):
    dataDict[data.iloc[i, 0]] = data.iloc[1, range(i, len(data.columns), 6)]

print(dataDict)
```

```python
name = input("행정 구역을 입력하세요: ")
index = 0
# dict 타입을 반복할때 변수를 두개 주면 키 밸류를 가져오게된다
for i, j in dataDict.items():
    if name in i:
        findData = dataDict.get(i)

print(pd.DataFrame(findData))

figure = plt.figure()
axes = figure.add_subplot(111)

axes.plot(findData)
# axes.pie(findData, dataDict.index.values)
plt.show()
```
